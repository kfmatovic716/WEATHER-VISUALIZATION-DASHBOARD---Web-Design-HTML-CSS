```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
from scipy import stats
plt.style.use('seaborn')

real_estate_path = "combined_clean.csv"

# Read the data 
neighborhood_data = pd.read_csv(real_estate_path)
neighborhood_data
```

```python
plt.figure(figsize=(15,10), facecolor="#9ecae1")
width_bar = 0.50
n_neighbh = 10
index_neighbh = np.arange(n_neighbh)
neighborhood_population = (40227, 46050, 61469, 53974, 23454, 33173, 43027, 35232, 34994, 55443)
sorted_neighbhlist = ["Downtown", "Encino", "Hollywood", "Mar Vista", "Pacific Palisades", "Silver Lake", "South Los Angeles", "Studio City", "Venice", "Westwood"]
plt.bar(index_neighbh, neighborhood_population, width_bar, edgecolor='black', label='Population', color='#fc8d62')
plt.xlabel("Neighborhood", fontsize=18)
plt.ylabel('Population', fontsize=18)
width_bar = 0.25
n_ptypes = 10
index_neighbh = np.arange(n_ptypes)
plt.title('Population by Neighborhood', fontsize=20, weight="bold")
plt.legend(fontsize=17, frameon=True, facecolor="#9ecae1")
plt.xticks(index_neighbh + width_bar, sorted_neighbhlist, rotation=90, fontsize=15)
plt.yticks(fontsize=15)
plt.savefig("neighborhood_population.png")
plt.show()
```
